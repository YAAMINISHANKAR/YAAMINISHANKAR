```python
import numpy as npy
import pandas as pds
import matplotlib.pyplot as pylt
%matplotlib inline
import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import squareform, pdist

a = npy.array([0.4005,0.2148,0.3457,0.2652,0.0789,0.4548])
b = npy.array([0.5306,0.3854,0.3156,0.1875,0.4139,0.3022])

point = ['P1','P2','P3','P4','P5','P6']
data = pds.DataFrame({'Point':point, 'x cordinate':a, 'y cordinate':b})
data = data.set_index('Point')
data
```

```python
dist = pds.DataFrame(squareform(npy.round(pdist(data[['x cordinate', 'y cordinate']]),4), 'euclidean'), columns=data.index.values, index=data.index.values)
dist
```

```python
pylt.figure(figsize=(10,4)) 
pylt.title("Dendrogram with Single inkage")  
dend = sch.dendrogram(sch.linkage(data[['x cordinate', 'y cordinate']], method='single'), labels=data.index)
```

```python
pylt.figure(figsize=(10,4)) 
pylt.title("Dendrogram with Complete inkage")  
dend = sch.dendrogram(sch.linkage(data[['x cordinate', 'y cordinate']], method='complete'), labels=data.index)
```

```python
pylt.figure(figsize=(10,4)) 
pylt.title("Dendrogram with Average inkage")  
dend = sch.dendrogram(sch.linkage(data[['x cordinate', 'y cordinate']], method='average'), labels=data.index)
```

```python
## 2) Use CC_GENERAL.csv given in the folder and apply:
# a) Preprocess the data by removing the categorical column and filling the missing values.
# b) Apply StandardScaler() and normalize() functions to scale and normalize raw input data.
# c) Use PCA with K=2 to reduce the input dimensions to two features.
# d) Apply Agglomerative Clustering with k=2,3,4 and 5 on reduced features and visualize
# result for each k value using scatter plot.
# e) Evaluate different variations using Silhouette Scores and Visualize results with a bar chart.
```

```python
import numpy as npy
import pandas as pds
import seaborn as sbn
import matplotlib.pyplot as pylt
from sklearn import preprocessing,metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

import warnings
warnings.filterwarnings("ignore")
```

```python
dataframe = pds.read_csv('/bin/CC GENERAL.csv')
dataframe.info()
```

```python
dataframe.head()
```

```python
dataframe.describe()
```

```python
df = dataframe.drop(['CUST_ID'], axis=1)
df.head()
```

```python
df.isnull().any()
```

```python
df.fillna(dataframe.mean(), inplace=True)
df.isnull().any()
```

```python
df.corr().style.background_gradient(cmap="Greens")
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]


scaler = preprocessing.StandardScaler()
scaler.fit(x)
X_scaled_array = scaler.transform(x)
X_scaled_df = pds.DataFrame(X_scaled_array, columns = x.columns)
```

```python
#Normalization is the process of scaling individual samples to have unit norm. 
#This process can be useful if you plan to use a quadratic form such as the dot-product or any other kernel to quantify the similarity of any pair of samples.
X_normalized = preprocessing.normalize(X_scaled_df)
# Converting the numpy array into a pandas DataFrame
X_normalized = pds.DataFrame(X_normalized)
```

```python
pca2 = PCA(n_components=2)
principalComponents = pca2.fit_transform(X_normalized)

principalDf = pds.DataFrame(data = principalComponents, columns = ['P1', 'P2'])

finalDf = pds.concat([principalDf, df[['TENURE']]], axis = 1)
finalDf.head()
```

```python
pylt.figure(figsize=(7,7))
pylt.scatter(finalDf['P1'],finalDf['P2'],c=finalDf['TENURE'],cmap='prism', s =5)
pylt.xlabel('pc1')
pylt.ylabel('pc2')
```

```python
ac2 = AgglomerativeClustering(n_clusters = 2)
 
# Visualizing the clustering
pylt.figure(figsize =(6, 6))
pylt.scatter(principalDf['P1'], principalDf['P2'],
           c = ac2.fit_predict(principalDf), cmap ='rainbow')
pylt.show()
```

```python
ac3 = AgglomerativeClustering(n_clusters = 3)
 
# Visualizing the clustering
pylt.figure(figsize =(6, 6))
pylt.scatter(principalDf['P1'], principalDf['P2'],
           c = ac3.fit_predict(principalDf), cmap ='rainbow')
pylt.show()
```

```python
ac4 = AgglomerativeClustering(n_clusters = 4)
 
# Visualizing the clustering
pylt.figure(figsize =(6, 6))
pylt.scatter(principalDf['P1'], principalDf['P2'],
           c = ac4.fit_predict(principalDf), cmap ='rainbow')
pylt.show()
```

```python
ac5 = AgglomerativeClustering(n_clusters = 5)
 
# Visualizing the clustering
pylt.figure(figsize =(6, 6))
pylt.scatter(principalDf['P1'], principalDf['P2'],
           c = ac5.fit_predict(principalDf), cmap ='rainbow')
pylt.show()
```

```python
k = [2, 3, 4, 5]
 
# Appending the silhouette scores of the different models to the list
silhouette_scores = []
silhouette_scores.append(
        silhouette_score(principalDf, ac2.fit_predict(principalDf)))
silhouette_scores.append(
        silhouette_score(principalDf, ac3.fit_predict(principalDf)))
silhouette_scores.append(
        silhouette_score(principalDf, ac4.fit_predict(principalDf)))
silhouette_scores.append(
        silhouette_score(principalDf, ac5.fit_predict(principalDf)))

 
# Plotting a bar graph to compare the results
pylt.bar(k, silhouette_scores)
pylt.xlabel('Number of clusters', fontsize = 20)
pylt.ylabel('S(i)', fontsize = 20)
pylt.show()
```
